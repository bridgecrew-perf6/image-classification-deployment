```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import glob
import os
from keras.applications.vgg16 import VGG16
```

```python
from tensorflow.keras.applications import Xception,MobileNetV2
```

```python
print(os.listdir("images"))
size=256
```

```python
glob.glob("images/train/*/*.jpg")
```

```python
train_img=[]
train_label=[]
for image in glob.glob("images/train/*/*.jpg"):
    label=image.split("\\")[-2]
    img=cv2.imread(image)
    img=cv2.resize(img,(size,size))
    
    train_img.append(img)
    train_label.append(label)
```

```python
train_img=np.array(train_img)
train_label=np.array(train_label)
```

```python
test_img=[]
test_label=[]
for image in glob.glob("images/test/*/*.jpg"):
    label=image.split("\\")[-2]
    img=cv2.imread(image)
    img=cv2.resize(img,(size,size))
    
    test_img.append(img)
    test_label.append(label)
```

```python
test_img=np.array(test_img)
test_label_=np.array(test_label)
```

```python
train_img=train_img/255.0
test_img=test_img/255.0
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
train_label=le.fit_transform(train_label)
test_label=le.fit_transform(test_label_)
```

```python
X_train,y_train,X_test,y_test=train_img , train_label , test_img , test_label
```

```python
from tensorflow.keras.utils import to_categorical
one_hot_train=to_categorical(y_train)
one_hot_test=to_categorical(y_test)
```

```python
vgg=MobileNetV2(include_top=False,input_shape=(size,size,3))
```

```python
for i in vgg.layers:
    i.trainable=False
vgg.summary()
```

```python
feature_Extraction=vgg.predict(X_train)
```

```python
rf_X_train=feature_Extraction.reshape(feature_Extraction.shape[0],-1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()#n_estimators = 50, random_state = 42)
rf.fit(rf_X_train,y_train)
```

```python
feature_Extraction_X_test=vgg.predict(X_test)
rf_X_test=feature_Extraction_X_test.reshape(feature_Extraction_X_test.shape[0],-1)
```

```python
y_pred=rf.predict(rf_X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_pred,y_test),annot=True,cmap="coolwarm")
```
